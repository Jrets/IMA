### $P = 2x+2y = 144 =>$ $y = \frac{144-2x}{2} = 77 - x $ <br>  $S = x*y =  77x - x^2, x > 0 $ <br> $S^\prime = (77x - x^2)^\prime$ <br>$ (77x - x^2)^\prime = 0$ <br> $77-2x = 0$ <br> $ x = 38.5 $

### –¢–æ—á–∫–∞ x=38.5 —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—é –∑–∞–¥–∞–Ω–∏—è.  –ü—Ä–∏ x < 38.5,S‚Ä≤>0, –∞ –ø—Ä–∏ x>38.5,S‚Ä≤<0. –ó–Ω–∞—á–∏—Ç —Ç–æ—á–∫–∞ x=38.5  —è–≤–ª—è–µ—Ç—Å—è —Ç–æ—á–∫–æ–π –º–∞–∫—Å–∏–º—É–º–∞ —Ñ—É–Ω–∫—Ü–∏–∏. <br>$ y = 77 - x = 38.5 $ <br> –ü—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫ –±—É–¥–µ—Ç –∏–º–µ—Ç—å –Ω–∞–∏–±–æ–ª—å—à—É—é –ø–ª–æ—â–∞–¥—å, –µ—Å–ª–∏ —ç—Ç–æ –∫–≤–∞–¥—Ä–∞—Ç —Å–æ —Å—Ç–æ—Ä–æ–Ω–æ–π 38.5—Å–º.

#### –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è —ç–∫—Å—Ç—Ä–µ–º—É–º, –ª–æ–∫–∞–ª—å–Ω—ã–π –º–∏–Ω–∏–º—É–º –≤ —Ç–æ—á–∫–µ x=0

### $b) y = x^3 =>$  $ y ^\prime = (x^3)^\prime = 3x^2$ <br> $3x^2 = 0, x = 0, y = 0$ 

```python
x = np.linspace(-10, 10, 100)
y = x**3
plt.plot(x, y)
         
a = np.linspace(-10, 10, 100)
b = 3*a**2
plt.plot(a, b)
```

#### —Å–º–µ–Ω—ã –∑–Ω–∞–∫–∞ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π, –ø–æ—Å–ª–µ x = 0 –Ω–µ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç. –≠–∫—Å—Ç—Ä–µ–º—É–º–∞ –Ω–µ—Ç. 

### $c) y = e^{3x} =>$ $ y^\prime = (e^{3x})^\prime = 3e^{3x}$ <br> $3e^{3x} = 0$

```python
x = np.linspace(-10, 10, 500)
y = math.e**(3*x)
plt.plot(x, y)
```

#### —Ñ—É–Ω–∫—Ü–∏—è –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç, —ç–∫—Å—Ç—Ä–µ–º—É–º–∞ –Ω–µ—Ç

### $d) y = x^3 - 5x => y^\prime = (x^3 - 5x)^\prime = 3x^2 - 5 $ <br> $3x^2 - 5 = 0 => x = \sqrt[]{\frac {5}{3}}  $

```python
x = np.linspace(-10, 10, 100)
y = x**3 - 5*x
plt.plot(x, y)

a = np.linspace(-10, 10, 100)
b = 3*a**2 - 5
plt.plot(a, b)
```
