## Тема “Теоремы о пределах”


#### Найти предел:


$a)$ $\lim_{x \to 0} \frac{sin(2x)}{4x}$<br>
Используется свойство первого замечательного предела:<br>
$\lim_{x \to 0} \frac{sin(2x)}{4x}$ = $\frac{sin(2x)}{4x}$ = $\frac{sin(2x)}{2*2x}$ = $\frac{1}{2}$

$b)$ $\lim_{x \to 0} \frac{x}{sin(x)}$<br>
следствие первого замечательного предела<br>
$\lim_{x \to 0} \frac{x}{sin(x)}$ = $lim_{x \to 0} 1 / \frac{sin(x)}{x}$ = $1$

$c)$ $\lim_{x \to 0} \frac{x}{arcsin(x)}$<br>

Следствие из первого замечательного предела:<br>
$\lim_{x \to 0} \frac{x}{arcsin(x)}$ = $\lim_{x \to 0} 1 / \frac{arcsin(x)}{x}$ = 1

$d)$ $\lim_{x \to \infty} (\frac{4x+3}{4x-3})^{6x}$<br>
<br>
$\lim_{x \to \infty} (\frac{4x+3}{4x-3})^{6x}$ = $\lim_{x \to \infty} (\frac{4x-3+6}{4x-3})^{6x}$ = $\lim_{x \to \infty} (1 +\frac{6}{4x-3})^{\frac{4x-3}{6}{6x}\frac{6}{4x-3}}$ = <br> = $e^{\lim_{x \to \infty}{6x}\frac{6}{4x-3} }$ = $e^{\lim_{x \to \infty}\frac{36x}{4x-3}}$ = $e^{\frac{36}{4}} = e^9$

$e)$$\lim_{x \to \infty} \frac{sin(x) + ln(x)}{x}$ = $\lim_{x \to \infty} (\frac{sin(x)}{x}$ + $\frac{ln(x)}{x})$ = $0 + 0$ = $0$ <br>
