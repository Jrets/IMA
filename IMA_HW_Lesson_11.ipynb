## Тема 7 “Ряды”

### 7.1 Исследовать ряд на сходимость, используя признак д’Аламбера: <br> $\sum\limits_{n=1}^\infty\frac{n^n}{(n!)^2}$

## $\lim_{n \to \infty}\frac{(n+1)^{n+1}}{((n+1)!)^2}:\frac{n^n}{(n!)^2}$ = $\lim_{n \to \infty}\frac{(n+1)(n+1)^{n}(n!)^2}{((n+1)!)^2n^n}$ = $\lim_{n \to \infty}\frac{(n+1)(n+1)^{n}(n!)^2}{(n!(n+1))(n!(n+1))n^n}$ = <br>$\lim_{n \to \infty}\frac{(n+1)^{n}}{(n+1)n^n}$ = $\lim_{n \to \infty}\frac{(n+1)(n+1)^{n-1}}{(n+1)n^n}$ = $\lim_{n \to \infty}\frac{(n+1)^{n-1}}{n^n} = 0$ ряд сходится

### 7.2 Исследовать ряд на сходимость, используя радикальный признак Коши: <br> $\sum\limits_{n=1}^\infty\frac{n}{2^n}$

## $\lim_{n\to\infty}\sqrt[n]{\frac{n}{2^n}}$ = $\frac{\lim_{n\to\infty}\sqrt[n]{n}}{\lim_{n\to\infty}\sqrt[n]{2^n}}$ = $\frac{1}{2}$=> ряд сходитcя

### 7.3 Исследовать ряд на сходимость, используя признак Лейбница: <br> $\sum\limits_{n=1}^\infty\frac{(-1)^n}{n + \ln{n}}$

### $\lim_{n\to\infty}\frac{(-1)^n}{n + ln(n)}$ = $0$

### $\frac{1}{n + ln(n)} \geq \frac{1}{(n + 1)ln(n+1)}$ монотонно убывает по модулю, ряд сходится

### 7.4 Исследовать ряд на сходимость, используя признак Раабе: <br> $\sum\limits_{n=1}^\infty\frac{3^n}{2^n}$

### $D =\lim_{n\to\infty} (n(\frac{a(n)}{a(n+1)} - 1))$ => $\lim_{n\to\infty}  (n(\frac{3^n}{2^n}:\frac{3^{n+1}}{2^{n+1}} - 1))$ = $\lim_{n\to\infty}  (n(\frac{3^n}{2^n}*\frac{2^{n}2}{3^{n}3} - 1))$ = $n(\frac{2}{3} - 1)$ = $-\frac{1}{3}n$<br> ряд расходится
