#### 3.1 Даны 4 последовательности. Необходимо:
исследовать их на монотонность; \
исследовать на ограниченность; \
найти пятый по счету член.


```python
sousre_seq_1 = list(range(1,15,1))
sousre_seq_2 = list(range(2,15,1))
```

```python
def monotone(sequence):
    for i in range(1, len(sequence)):
        if sequence[i] < sequence[i+1]:
            return "Последовательность монотонно возрастает"
        elif sequence[i] > sequence[i+1]:
            return "Последовательность монотонно убывает"
        else:
            return "Последовательность монотонно не изменяется"
```

```python
equation_a = list(map(lambda n: 2**n - n, sousre_seq_1))
print(equation_a)
print(monotone(equation_a))
# последовательность ограничена "снизу" числом 1
print(f'Пятый член последовательности {equation_a[4]}')
```

```python
equation_b = list(map(lambda n: 1/(1 - n), sousre_seq_2))
print(equation_b)
print(monotone(equation_b))
# последовательность ограничена "снизу" числом -1
print(f'Пятый член последовательности {equation_b[4]}')
```

```python
equation_c = list(map(lambda n: -1**n +math.sqrt(2*n), sousre_seq_1))
print(equation_c)
print(monotone(equation_c))
# последовательность ограничена "снизу" числом 0
print(f'Пятый член последовательности {equation_c[4]}')
```

```python
equation_d = list(map(lambda n: -1**(2*n) +1/(n**2), sousre_seq_1))
print(equation_d)
print(monotone(equation_d))
# последовательность ограничена сверху числом 2
print(f'Пятый член последовательности {equation_d[4]}')
```

#### 3.3*На языке Python предложить алгоритм вычисляющий численно предел с точностью e=10^(-7)

```python
import sympy 
```

```python
e = 10 ** (-7)
i = 1
n = 1
x_0 = n / pow(sympy.factorial(n), 1 / n)
while True:
    i += 1
    n += 1
    x_n = n / pow(sympy.factorial(n), 1 / n)
    if abs(x_0 - x_n) <= e:
        break
    x_0 = x_n
print(f'n_iteration: {i}')
print(f'lim = {x_0}')
```
